# Dataset Used in Project

This is the Star Wars Episode V script dataset that was used for this project.

Refer to the link: https://www.kaggle.com/datasets/xvivancos/star-wars-movie-scripts?select=SW_EpisodeV.txt

--------

After opening the file, the first line is skipped in the file. This line is "character" "dialogue" which is not necessary for this analysis.

Each letter in the script becomes lowercase.

A dictionary is created, splitting each line in the script by line number, name of character and dialogue. This dictionary is stored in a list.

Printing the list returns each individual line: line number, character name, and the dialogue by commas. Each line is then seaparated by curly brackets {}.

```python
#create list: separated by each line: line number, character  name, and dialogue
sw_dic = []
with open("SW_EpisodeV.txt", "r") as text:
    text = text.readlines()[1:]  #skip first line in file
    for line in text:
        line = line.lower().split('"') #lowercase all words and split by quotes
        dialogue = line.pop(5)
        name = line.pop(3)
        line_number = line.pop(1)
        new_obj = {
            "line": line_number,
            "name": name,
            "dialogue": dialogue
        }
        sw_dic.append(new_obj)
#print out each line separately by line number, character name, and dialogue
print(sw_dic)
```
